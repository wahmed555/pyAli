```python
# funtion for  generating global enumerated variable from list(l_M containing varaible names for df ) and 
# assign the the elements of l_M to the generated varable

def gen_enum():
    for n, val in enumerate(L_M,start=1):
        globals()["df%d"%n] = val
        print(["df%d"%n])
        
gen_enum()
```
